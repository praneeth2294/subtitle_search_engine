## **Step 5 - Unzipping the content of 385th row and decoding using `latin-1`**

```python
import zipfile
import io

# Assuming 'content' is the binary data from your database
binary_data = df.iloc[385, 2]

# Decompress the binary data using the zipfile module
with io.BytesIO(binary_data) as f:
    with zipfile.ZipFile(f, 'r') as zip_file:
        # Reading only one file in the ZIP archive
        subtitle_content = zip_file.read(zip_file.namelist()[0])

# Now 'subtitle_content' should contain the extracted subtitle content
print(subtitle_content.decode('latin-1'))  # Assuming the content is latin-1 encoded text
```

**Look's like it worked.**

## **Step 6 - Applying the above Function on the Entire Data**

```python
import zipfile
import io

count = 0

def decode_method(binary_data):
    global count
    # Decompress the binary data using the zipfile module
    # print(count, end=" ")
    count += 1
    with io.BytesIO(binary_data) as f:
        with zipfile.ZipFile(f, 'r') as zip_file:
            # Assuming there's only one file in the ZIP archive
            subtitle_content = zip_file.read(zip_file.namelist()[0])

    # Now 'subtitle_content' should contain the extracted subtitle content
    return subtitle_content.decode('latin-1')  # Assuming the content is UTF-8 encoded text
```

```python
df['file_content'] = df['content'].apply(decode_method)

df.head()
```

```python
import re
```

```python
# ----------------- 🔹 Text Cleaning Function 🔹 -----------------
def clean_subtitles(text, index):
    if pd.isna(text):
        return ""

    # Remove timestamps (e.g., "00:00:06,000 --> 00:00:12,074")
    text = re.sub(r"\d{2}:\d{2}:\d{2},\d{3} --> \d{2}:\d{2}:\d{2},\d{3}", "", text)

    # Remove subtitle numbering (e.g., "1", "2", "3" on separate lines)
    text = re.sub(r"^\d+\s*$", "", text, flags=re.MULTILINE)

    # Normalize newlines and remove excessive `\r\n`
    text = text.replace("\r", "").replace("\n", " ")

    # Remove unnecessary special characters (but keep punctuation)
    text = re.sub(r"[^\w\s.,!?'-]", "", text)

    # Reduce multiple spaces into a single space
    text = re.sub(r"\s+", " ", text).strip()

    # 🚀 **Remove unwanted text in the first row**
    if index == 0:
        text = re.sub(r"Watch any video online with Open[-\s]*SUBTITLES Free Browser extension osdb\.link[/]*ext", "", text, flags=re.IGNORECASE).strip()

    return text

# ----------------- 🔹 Load and Process Subtitles 🔹 -----------------

def load_and_process_subtitles(file_path):
    df = pd.read_csv(file_path)  # Update with your subtitle file
    df["cleaned_subtitles"] = df.apply(lambda row: clean_subtitles(row["file_content"], row.name), axis=1)
    return df

subtitle_data_path = "/content/drive/MyDrive/shazam/cleaned_subtitles1.csv" ## pasted the csv file link which i saved in mydrive
df = load_and_process_subtitles(subtitle_data_path)
```
